```python
import pandas as pd
import plotly.graph_objects as go
```

```python
def import_data(file_name):
  df = pd.read_csv(file_name)
  return df
 
df = import_data('BTC-USD.csv')
df['Date'] = pd.to_datetime(df['Date'])   
df = df.set_index('Date')    
df.head()
```

## Japanese Candlesticks

```python
df2 = df
# Last 3 Months
fig = go.Figure(data=[go.Candlestick(   
                      x = df2.index,                  
                      open=df2['Open'],
                      high=df2['High'],
                      low=df2['Low'],
                      close=df2['Close'])]
                )
fig.show()
```

### Some Observations : 

1. March 14 2021 - A <b>bearish Marubozu</b> candle is observed, so we expect prices to fall(downtrend)  as selling pressure increases . As you can see , next day we get new low closing prices.   

2. March 25, 2021 - A <b>spinning top</b> candle is observed, so anything can happen, Either continuation of trend (i.e. downtrend here) or reversal (i.e. bulls take over the market) . 

3. March 28, 2021 - A <b> Dozi</b> candle is observed, so market is indecisive.

Note - In case of spinning top trend reversal occurred but not in case of Dozi . That's why we said anything can happen in case of Dozi or Spinning top candle sticks. 

4.  March 26, 2021 - A <b>bullish Marubozu</b> candle is observed, so we expect uptrend (i.e. increase in prices ). From chart, it is evident that next day's closing prices is higher than today's opening prices. 


5.  April 25, 2021 - A <b>Umbrella</b> candle is observed . As the ratio of lower shadow to real body is  ≈ 2  and  it is infact a <b>Hammer</b> as it follows prior trend rule(downtrend). So we expect reversal(uptrend).

6. May 9, 2021 - A <b> Umbrella</b> candle is observed. As ratio of lower shadow to real body is greater than 2 and it is infact a <b>Hanging man</b> as it follows prior trend rule(uptrend). So we expect reversal(downtrend).

### Multiple CandleSticks Pattern 

1. 13 March - 14 March, A <b>Dark Cloud Cover</b> is formed with engulfed area of almost 50%. 

2. 21 Feb - 22 Feb , A <b>bearish engulfing</b> pattern is observed . Infact bear didn't let Day-2 price open higher than Day-1 Close price  and also low of Day-2 is very low compared to opening price of Day-1 indicating bears are dominanat now.
